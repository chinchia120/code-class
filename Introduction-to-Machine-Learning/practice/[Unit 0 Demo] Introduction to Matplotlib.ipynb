# **Matplotlib介紹**
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.
* Create publication quality plots.
* Make interactive figures that can zoom, pan, update.
* Customize visual style and layout.
* Export to many file formats.

For this tutorial, we start by importing the package.

```python
import matplotlib.pyplot as plt
```

#### **Note 1. 畫折線圖**

```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]    # 年份
pop = [2.519, 3.692, 5.263, 6.972]   # 全球總人口（單位：10億)
plt.plot(year, pop)       # 畫折線圖
plt.show()           # year:x軸，pop:y軸# 顯示出折線圖
```

#### **Note 2. 畫點散圖**

```python
import matplotlib.pyplot as plt
plt.scatter(year, pop)
plt.show()
```

**畫點散圖(進階版)**

```python
import numpy as np
import matplotlib.pyplot as plt
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N))**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

#### **Note 3.畫直方圖**

```python
import numpy as np
import matplotlib.pyplot as plt
# 資料樣本為1000個身高模擬資料
values = np.round(np.random.normal(1.75, 0.20, 1000), 2)
# bins表示直方圖劃分的區間數
plt.hist(values, bins=10)
plt.show()
```

#### **Note 5. 一個視窗繪製多個圖表**

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2, figsize=(8, 4))

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

# 第一種方法
ax[0, 0].plot(year, pop)
ax[0, 1].scatter(year, pop)

# 第二種方法
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N))**2
subfig = plt.subplot(2, 2, 3)
subfig.scatter(x, y, s=area, c=colors, alpha=0.5)
subfig = plt.subplot(2, 2, 4)
subfig.scatter(y, x, s=area, c=colors, alpha=0.5)
fig.suptitle('Scatter with regression')
plt.show()
```
